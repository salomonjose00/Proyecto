**4.4.** En la siguiente regresión simple, MonthlyIncome=  b0  + b1*YearsAtCompany, ¿Cuál sería el valor de Durbin-Watson? ¿Qué significa esto?

**(MonthlyIncome y YearsAtCompany)**

```python
#MonthlyIncome = YearsAtCompany*b1 + b0

model_ols = sm.OLS.from_formula("MonthlyIncome ~ YearsAtCompany", data=df)
result = model_ols.fit()
result.summary()
```

**Answer 4.4** 

**El valor de Durbin-Watson es de 2.034.**

**Como en este caso el estadístico es mayor que 2 entonces no hay correlación serial positiva entre las variables de MonthlyIncome y YearsAtCompany**.

**4.6.** Obtener el número de tarifa diaria que se genera con empleados, según el departamento al que pertenecen. ¿Cuál departamento genera más tarifa diaria?

**(Department y DailyRate)**

```python
import plotly.express as px
```

```python
datosdd = df.groupby('Department')['DailyRate'].sum().reset_index()
datosdd
```

**4.7.** ¿La tarifa diaria del empleado se ve afectada por el total de años trabajados?

**(DailyRate y TotalWorkingYears)**

```python
bdata= df.groupby(['TotalWorkingYears'])['DailyRate'].sum().reset_index()
bdata.head()
```

```python
fig = px.bar(bdata,
             x= 'TotalWorkingYears',
             y= 'DailyRate',
             title= 'Número de años en relación con tarifa diaria')
fig.show()
```

**Answer 4.7** 

**Con el gráfico anterior podemos concluir que la tarifa diaria del empleado si se ve afectada con el paso de los años. Con el paso de los años la tarifa va disminuyendo**
